```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import linear_model

df = pd.read_csv("data.csv")
print(df)
```

```python
%matplotlib inline
plt.xlabel("areas")
plt.ylabel("prices")
plt.scatter(df.area, df.price,color ="red",marker="*")
```

```python
reg = linear_model.LinearRegression()
reg.fit(df[['area']],df.price)
```

```python
x = 1000

predicted_price = reg.predict([[x]])  
print(f"Predicted price for {x} sq ft: {predicted_price[0]}")

plt.scatter(df.area, df.price, color="red", marker="*")  
plt.plot(df.area, reg.predict(df[['area']]), color='blue')
plt.title("Area vs Price Prediction")
plt.xlabel("Area (sq ft)")
plt.ylabel("Price (USD)")
plt.show()


```

```python
d = pd.read_csv("New.csv")
d.head(3)
print(d)
p = reg.predict(d)
print(p)
d['predicted price'] = p
d.to_csv("New2.csv",index=False)
```
