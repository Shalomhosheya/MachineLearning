```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model

df = pd.read_csv("data2.csv")
print(df)
```

```python
plt.xlabel("year")
plt.ylabel("per capita income (US$)")
plt.scatter(df.year, df.income,color ="red",marker="*")
```

```python
reg = linear_model.LinearRegression()
reg.fit(df[['year']], df.income)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Assuming reg is a trained LinearRegression model
# Example training setup (uncomment if needed):
# training_data = pd.read_csv("training_data.csv")  # Replace with your actual training data file
# reg = LinearRegression()
# reg.fit(training_data[['year']], training_data['income'])

# Prediction for a specific year
predictedYear = input("year:")
predictedYear = int(predictedYear)
predictedIncome = reg.predict([[predictedYear]])[0]  # Extract the scalar value
print(f"Prediction for year {predictedYear}: ${predictedIncome}")

# Read the existing CSV file or create a new DataFrame if it doesn't exist
try:
    df = pd.read_csv("prediction_yearIncome.csv")
except FileNotFoundError:
    print("File not found. Creating a new DataFrame.")
    df = pd.DataFrame(columns=["year", "income"])

# Append the new prediction
df = pd.concat([df, pd.DataFrame({"year": [predictedYear], "income": [predictedIncome]})], ignore_index=True)

# Save the updated DataFrame back to the CSV file
df.to_csv("prediction_yearIncome.csv", index=False)

# Visualization
plt.scatter(df['year'], df['income'], color="red", marker="*")
plt.plot(df['year'], reg.predict(df[['year']]), color='blue')  # Regression line
plt.title("Year vs Income Prediction")
plt.xlabel("Year")
plt.ylabel("Per Capita Income (US$)")
plt.show()
```
